#Assignment

```python
#Single Assignment
beratbadansiswa1 = 90
print(beratbadansiswa1)
```

```python
#Multiple Assignment
nilai_kka_dias, nilai_kka_azka = 99, 98

print(nilai_kka_azka)
print(nilai_kka_dias)
```

#Operasi pada Python

```python
#Tanpa Variabel
print(2+8)
```

```python
#Operand disimpan ke vavriabel
kelereng_dias = 2
kelereng_azka = 8
print(kelereng_dias + kelereng_azka)
```

```python
#Hasil operasi disimpan ke variabel
total_kelereng = 8+2
print(total_kelereng)
```

```python
#Operand dan hasil operasi disimpan ke variabel
kelereng_azka = 8
kelereng_dias = 2
total_kelereng = kelereng_azka + kelereng_dias
print(total_kelereng)
```
